# Matplotlib 

## Introduction
Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating **scientific figures**. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically.
The official Matplotlib web page: http://matplotlib.org/

# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure. 

So, we could have two size plots embedded together.

- figure object
    - fig = plt.figure(): create a figure object
- axes object (can have mulitple using subplot - manugally create axes)
    - fig.add_axes([x0, y0, width, height]): create axes object from lower left point
    - axes.plot(x,y,color,label,..)
    - axes.set_xlabel
    - axes.set_ylabel
    - axes.set_title
    
## Legends, labels and titles
Decorate a figure with titles, axis labels, and legends.

1. **Figure titles**: A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance
2. **Axis labels**:  with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes
3. **Legends"**: **axex.legend()**, put **label=** arguments when plotting.
    - use **label="label text"** keyword argument in axex.plot() or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure.
    - specify the location: ax.legend(loc =0), 0 indicates the best location
    
    
## Setting colors, linewidths, linetypes

Many options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake): MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

- Color param: ax.plot(color = )
    - can put a color string, a RBG code(goolge #RGD HEX Code), and optionally provide an alpha value using the `color` and `alpha` keyword arguments to indicate opacity.
    
- linewidth (lw) param: ax.plot(linwidth=2)
- linestyle (ls) : linestyle = '--' dashed line
    - '-.' dash points
    - ':' small dash line
    
- marker: mark each element in the x array
    - marker = # possible marker symbols '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
    - markersize: to adjust marker size
    - markerfacecolor = 'yellow'
    - markeredgewidth
    - markeredgecolor

- set_xlim([lower, upper]): set a range for x axis
- set_ylim([lower, upper]): set a range for y axis
- `axis('tight')` for automatically getting "tightly fitted" axes ranges:


## Subplots()

- The **plt.subplots()** object will act as a more automatic **axis manager**.
    - fig, axes = **plt.subplots(nrow, ncol, plot #)**
    - When create multiple subplots, we need to iterate axes object OR (axes is a nd array) we could use index to set each axes

- A common issue with matplolib is overlapping subplots or figures. 
    - Deal with overlap: Use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content.

## Figure size, aspect ratio and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the **Figure** object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

- fig = plt.figure(figsize(width, height), dpi=)

## Saving figures
Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF. 

To save a figure to a file we can use the `savefig` method in the `Figure` class, we can also optionally specify the DPI and choose between different output formats:
- fig.savefig("name.png")
- fig.savefig("name.png", dpi=200)



# Special Plot Types

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x,y, 'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('String Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

```

```python
fix, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

#Display the figure object
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
fig
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y,'g')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'r')
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize = (3,2), dpi = 100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,x**2, label ='X Squared')
ax.plot(x,x**3, label= 'X Cubed')
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.legend(loc=0)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,2))
axes[0,1].plot(x,y)
axes[1,1].plot(y,x)
plt.tight_layout()
```
