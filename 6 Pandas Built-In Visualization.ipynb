# Plot Types
- df1['A'].plot(kind='hist', bins=30) OR df1['A'].plot.hist()/plot.box() ...

## df.plot.area     
- df2['col'].plot.area(alpha=0.4): area plot

## df.plot.barh    


## df.plot.hist: recommended
- df1['A'].plot.hist(bins=50)

## df.plot.line: ecommended    
- df1.plot.line(y='B',figsize=(12,3),lw=1): automatically takes df.index as x

## df.plot.scatter
- df1.plot.scatter(x='col1',y='col2', c='col3'): c means color
- df1.plot.scatter(x='col1',y='col',s=df1['C']*200): s is size based off another column. s parameter needs to be an array, not just the name of a column:

## df.plot.bar      
- df['col'].plot.bar(stacked=True): bar plot

## df.plot.box
- df2.plot.box() # Can also pass a by= argument for groupby
- df2['col1'].plot.box()

## df.plot.hexbin  
Useful for Bivariate Data, alternative to scatterplot, dot show in hex shape
- df.plot.hexbin(x='col1',y='col',gridsize=25,cmap='Oranges')

## df.plot.kde      
- df2['col1'].plot.kde()

## df.plot.density: same as kde
- df2['col1'].plot.density()
- df2.plot.density(): multiple density lines of each variable in data frame

## df.plot.pie

Advanced plot recommend use Seaborn.

### plot the df in a certain range of rows: use ix
- df3.ix[0:30].plot.area(alpha=0.4)


# Style Sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that your plots follow. I recommend using them, they make all your plots have the same look and feel more professional. You can even create your own if you want your company's plots to all have the same look (it is a bit tedious to create on though).

Here is how to use them.

**Use plt.style.use() before plot**
- plt.style.use('dark_background')
